```python
# Install dependencies (if not already installed)
!pip install torch transformers accelerate peft

from transformers import pipeline
import torch

# Initialize the pipeline with your model and load on GPU
pipe = pipeline("text-generation", model="meta-llama/Llama-2-7b-chat-hf", device=0 if torch.cuda.is_available() else -1)
device = 0 if torch.cuda.is_available() else -1
print(device)


```

```python


# Initialize the conversation and context
print("Chatbot: Hello! How can I assist you today?")
conversation_history = []

# Main conversational loop
while True:
    # Get user input
    user_input = input("You: ")

    # if the user wants to exit, break the loop
    if user_input.lower() in ["exit", "quit", "bye"]:
        print("Chatbot: Goodbye!")
        break

    # Append the new input to the conversation history
    conversation_history.append(f"You: {user_input}")

    # Create the full conversation context for the model
    context = "\n".join(conversation_history[-5:])  # Limit the context to the last 5 interactions for efficiency

    # Generate a response using the pipeline with max_new_tokens
    response = pipe(context,
                    max_new_tokens=50,   # Specify the maximum number of tokens to generate
                    do_sample=True,
                    top_p=0.85,        # Reduce top_p for more focused responses
                    temperature=0.6,  # Lower temperature for more determinism
                    truncation=True,
                    repetition_penalty=1.15
                    )

    # Extract the portion after "Assistant:" and print it
    generated_text = response[0]['generated_text']
    print(generated_text)

    if "Assistant:" in generated_text:
        chatbot_response = generated_text.split("Assistant:")[-1].strip()  # Extract the model's response part after "Assistant:"
    else:
        chatbot_response = generated_text.strip()  # Fallback in case "Assistant:" is missing

    # print(f"Chatbot: {chatbot_response}")

    # Append the chatbot's response to the conversation history
    conversation_history.append(f"Chatbot: {chatbot_response}")
```
