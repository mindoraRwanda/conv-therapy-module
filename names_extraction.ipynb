```python
import spacy

# Load SpaCy model
nlp = spacy.load('en_core_web_sm')

def extract_names(text):
    doc = nlp(text)
    names = [ent.text for ent in doc.ents if ent.label_ == 'PERSON']
    return names

```

```python
import pandas as pd
datasetx = pd.read_csv('data_csv/aggr_data.csv')

```

```python
# combine dataset into one string, dataset has two colums, "Context" and "Response"
dataset = datasetx['Context'] + datasetx['Response']

# convert the series into a list
dataset = dataset.tolist()
```

```python
# convert the list into a string
dataset = ' '.join(dataset)
```

```python
# extract names from the dataset, Text of length is 318727612,
# split into parts of 1M characters each
names_list = []
for i in range(0, len(dataset), 1000000):
    names = extract_names(dataset[i:i+1000000])
    names_list.append(names)
```

```python
print(len(names_list))
print(names_list[0])
```

```python
# flatten the list of lists into a single list
names_flat = []
for list in names_list:
    for name in list:
        names_flat.append(name)
```

```python
# remove duplicates
len(names_flat)
name_set = set(names_flat)
len(name_set)
```

```python
# remove words which start with a lower case letter
names_upper = [name for name in name_set if name[0].isupper()]

len(names_upper)
```

```python
# sort the names in alphabetical order
names_upper.sort()
```

```python
# save the names to a file 
with open('names.txt', 'w', encoding='utf-8') as f:
    for name in names_upper:
        f.write(name + '\n')
```
